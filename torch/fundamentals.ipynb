```python
import torch
print(torch. __version__)
```

## Scaler

```python
scaler=torch.tensor(45)
scaler.ndim  # number of dimensions 
scaler.shape # number of elements
print(scaler)
```

## Vector

```python
vector=torch.tensor([34,45,4])
vector.ndim  # number of dimensions 
vector.shape # number of elements
print(vector)
```

# MATRIX

```python
matrix=torch.tensor([[3,4],[6,7],[2,1]])
print(matrix.ndim)
print(type(matrix))
print(matrix)
print(matrix[0]) 
print(matrix[1])
print(matrix.shape) # 3 x 2 matrix
```
